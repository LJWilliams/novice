## Block example

```python
blocks = novice.open("block.png")
blocks
```

```python
blocks.inflation = 10
blocks
```

```python
blocks.size
```

```python
blocks[1:5, 1:5] = novice.YELLOW
blocks
```

```python
blocks.inflation = 25
blocks
```

```python
blocks[:, 0]                 = novice.VIOLET
blocks[:, blocks.height - 1] = novice.VIOLET
blocks[0, :]                 = novice.VIOLET
blocks[blocks.width - 1, :]  = novice.VIOLET
blocks
```
